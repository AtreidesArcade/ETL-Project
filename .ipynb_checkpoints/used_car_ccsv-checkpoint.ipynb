```python
# Import Dependencies
import pandas as pd
import csv
import pymongo
import requests
from splinter import Browser
from bs4 import BeautifulSoup
from webdriver_manager.chrome import ChromeDriverManager
```

```python
# path / import file
audi_df = pd.read_csv('Resources/audi_reviews.csv')
audi_df.head()
```

```python
#drop nulls
audi_df = audi_df.dropna()
```

```python
#set new index id 
audi_df = audi_df.set_index('Index_Id') 
```

```python
#create new vehicle naming convention from vehicle title field
vehicle_titles = audi_df['Vehicle_Title']

#initialize vehicle variable
Vehicle_Name = []

for vehicle in vehicle_titles:
    title_list = vehicle.split()
    name = pd.Series([title_list[0], title_list[1], title_list[2]])
    vehicle_name = name.str.cat(sep = '-')
    
    Vehicle_Name.append(vehicle_name)

audi_df['Vehicle_Name'] = Vehicle_Name

audi_df
```

```python
#grab/ group by

ratings = audi_df[['Rating','Vehicle_Name']]
ratings_groupby = ratings.groupby('Vehicle_Name').mean()

ratings_groupby
```

```python
# setup splinter
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
#initalize empty variables to hold returns
CarNames = []
CarPrices = []
InventoryType = []
Dealerships = []
Dealership_Distances = []

for x in range(1, 6):
    
    search = pd.Series(['https://www.cars.com/shopping/results/?page=',str(x),'&page_size=50&dealer_id=&keyword=&list_price_max=&list_price_min=&makes[]=audi&maximum_distance=200&mileage_max=&sort=best_match_desc&stock_type=new&year_max=&year_min=&zip=30341'])
    url = search.str.cat()
    browser.visit(url)
    
    html = browser.html
    soup = BeautifulSoup(html, 'html.parser')
    
    # grab vehicle details from results
    results = soup.find_all('div', class_='vehicle-details')
    
    for result in results:
        # set variables to hold element / class results
        name = result.find('h2', class_='title').text
        price = result.find('span', class_='primary-price').text
        dealer = result.find('div', class_='dealer-name').strong.text
        distance = result.find('div', class_='miles-from').text
        inventory = result.find('p', class_='stock-type').text
    
    
        # append empty variables to list
        CarNames.append(name)
        CarPrices.append(price)
        Dealerships.append(dealer)
        Dealership_Distances.append(distance)
        InventoryType.append(inventory)
```

```python
# create dictionary to hold lists of all results new scraped

car_dict = {
    'Name': CarNames,
    'Price($)': CarPrices,
    'InventoryType': InventoryType,
    'Distance': Dealership_Distances,
    'Dealership': Dealerships
}

car_df = pd.DataFrame(data = car_dict)
car_df.head()
```

```python
vehicle_titles = car_df['Name']

#new car unique vehicle name.
Vehicle_Name = []

for vehicle in vehicle_titles:
    
    
    title_list = vehicle.split()
    name = pd.Series([title_list[0], title_list[1], title_list[2]])
    vehicle_name = name.str.cat(sep = '-')
    
    Vehicle_Name.append(vehicle_name)

car_df['Vehicle_Name'] = Vehicle_Name
car_df
```

```python
#create list to hold vehicle details for all used audi
CarNames = []
CarPrices = []
InventoryType = []
Dealerships = []
Dealership_Distances = []

#loopy loop
for x in range(1, 6):
    
    search = pd.Series(['https://www.cars.com/shopping/results/?page=',str(x),'&page_size=50&dealer_id=&keyword=&list_price_max=&list_price_min=&makes[]=audi&maximum_distance=200&mileage_max=&sort=best_match_desc&stock_type=used&year_max=&year_min=&zip=30341'])
    url = search.str.cat()
    browser.visit(url)
    
    html = browser.html
    soup = BeautifulSoup(html, 'html.parser')
    
    # grab page elements with results
    results = soup.find_all('div', class_='vehicle-details')
    
    for result in results:
    
        # initalize empty variables with results
        name = result.find('h2', class_='title').text
        price = result.find('span', class_='primary-price').text
        dealer = result.find('div', class_='dealer-name').strong.text
        distance = result.find('div', class_='miles-from').text
        inventory = result.find('p', class_='stock-type').text
        
        # append variables with vehicle details
        CarNames.append(name)
        CarPrices.append(price)
        Dealerships.append(dealer)
        Dealership_Distances.append(distance)
        InventoryType.append(inventory)
```

```python
used_dict = {
    'Name': CarNames,
    'Price($)': CarPrices,
    'InventoryType': InventoryType,
    'Distance': Dealership_Distances,
    'Dealership': Dealerships
}

used_df = pd.DataFrame(data = used_dict)

browser.quit()
```
